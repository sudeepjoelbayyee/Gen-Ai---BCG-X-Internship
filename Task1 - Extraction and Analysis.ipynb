```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Financial_report.csv")
df
```

### Analyzing Financial Data

#### Calculate Year-over-Year Growth Rates

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
df
```

#### Analyze and Visualize

```python
# Group by company and calculate average growth rates
grouped_data = df.groupby('Company').mean()

# Print the average growth rates
print("Average Growth Rates:")
print(grouped_data[['Revenue Growth (%)', 'Net Income Growth (%)']])

# Visualize revenue growth
plt.figure(figsize=(10, 6))
plt.bar(df['Company'], df['Revenue Growth (%)'])
plt.title("Revenue Growth by Company")
plt.xlabel("Company")
plt.ylabel("Growth Rate (%)")
plt.show()
```

### Analyzing Profitability and Financial Health

#### Calculate Profitability Ratios

```python
df['Profit Margin (%)'] = (df['Net Income'] / df['Total Revenue']) * 100
```

```python
df
```

#### Visualize profit margins

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Company'], df['Profit Margin (%)'])
plt.title("Profit Margin by Company")
plt.xlabel("Company")
plt.ylabel("Profit Margin (%)")
plt.show()
```

### Analyzing Cash Flow

```python
df
```

#### Calculate Cash Flow Ratios

```python
df['Cash Flow from Operations Ratio'] = df['Cash Flow from Operating Activities'] / df['Total Revenue']
```

```python
df
```

#### Calculate Average Growth Rates

```python
# Calculate average growth rates
grouped_data = df.groupby('Company').mean()

print("Average Growth Rates:")
print(grouped_data[['Revenue Growth (%)', 'Net Income Growth (%)']])
```

```python
df
```

```python
df['Cash Flow from Operations Growth (%)'] = df.groupby(['Company'])['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
df
```

```python
df['Assets Growth (%)'] = df.groupby(['Company'])['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby(['Company'])['Total Liabilities'].pct_change() * 100
```

```python
df
```

#### Calculating average Growth rates

```python
# Calculate average growth rates for each company
average_growth_rates = df.groupby('Company').mean()

print("Average Growth Rates:")
print(average_growth_rates[['Revenue Growth (%)', 'Net Income Growth (%)', 'Assets Growth (%)', 'Liabilities Growth (%)', 'Cash Flow from Operations Growth (%)']])
```

```python
average_growth_rates[['Revenue Growth (%)', 'Net Income Growth (%)', 'Assets Growth (%)', 'Liabilities Growth (%)', 'Cash Flow from Operations Growth (%)']].to_csv("summary_financial_analysis.csv",index=False)
```
